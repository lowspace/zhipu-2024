```python
import sys
import os
import json
import time
import re

cwd = os.getcwd()
os.chdir(cwd)
sys.path.append('tools')

import chat
import parse_data
```

```python
def parse_database_and_table(query: str) -> dict:
    """
    Parse the given SQL query to return the database and table names in a dictionary.
    
    Args:
    - query (str): The SQL query to parse.
    
    Returns:
    - dict: A dictionary with 'database' and 'table' keys.
    """

    pattern = r'FROM\s+([a-zA-Z0-9_]+)\.([a-zA-Z0-9_]+)'
    match = re.search(pattern, query, re.IGNORECASE)
    
    if match:
        database = match.group(1)
        table = match.group(2)
        return {'database': database, 'table': table}
    
    return {}

fpath = os.path.join(cwd, 'answer_tmp' + os.sep + 'stage_1-glm_4_plus-ner-v2.0.0-sql-HF.json')

data = parse_data.read_json(fpath)

for i in data[:]:
    ner = i['ner']['stage_1']['sql']

    sql_res = [j for k, v in ner.items() for j in v]
    cnt = 0
    tmp = []

    for j in sql_res:
        if j['result']:
            tmp.append(parse_database_and_table(j['query']))
            cnt += 1

    if cnt >= 2:
        print(i['tid'], i['team'][0]['question'])
        print(tmp, '\n')
```
