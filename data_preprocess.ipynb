# 验证 AStockBasicInfoDB.LC_StockArchives 中 {X}StockCode 的数字模式

沪市A股票买卖的代码六位数，是以600、601或603开头；B股买卖的代码是六位数，以900开头； 深市A股票买卖的代码是六位数，以000开头，B股买卖的代码也是六位数，以200开头； 沪市新股申购的代码是六位数，以730开头，深市新股申购的代码与深市股票买卖代码一样； 配股代码是六位数，沪市以700开头，深市以080开头； 中小板股票代码是六位数，以002开头； 创业板股票代码是六位数，以300开头。

REF: https://www.nomuraoi-sec.com/pages/investorEducation/educationList.jsp?sectionKeyType=4#:~:text=%E7%AD%94%EF%BC%9A%E6%B2%AA%E5%B8%82A%E8%82%A1%E7%A5%A8,700%E5%BC%80%E5%A4%B4%EF%BC%8C%E6%B7%B1%E5%B8%82%E4%BB%A5

```python
import requests
import json

url = "https://comm.chatglm.cn/finglm2/api/query"
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {sql_api}"
}
data = {
  "sql": "SELECT CASE WHEN SUBSTRING(AStockCode, 1, 3) = '600' THEN 'Starts with 600' WHEN SUBSTRING(AStockCode, 1, 3) = '601' THEN 'Starts with 601' WHEN SUBSTRING(AStockCode, 1, 3) = '603' THEN 'Starts with 603' ELSE 'Other' END as AStockCode_Pattern, COUNT(*) as Count FROM AStockBasicInfoDB.LC_StockArchives GROUP BY AStockCode_Pattern ORDER BY Count DESC;",
  "limit": 1000
}

response = requests.post(url, headers=headers, json=data)
print(json.dumps(response.json(), indent=2, ensure_ascii=False))
```

```python
from collections import Counter

data = [i['AStockCode'] for i in response.json()['data']]

prefix_cnt = Counter()

for i in data:
    prefix = i[:3]
    prefix_cnt[prefix] += 1

prefix_cnt
```
