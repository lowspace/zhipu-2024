```python
import requests

def execute_https_api_request(api_url, api_key, query):
    """
    Executes an HTTPS API request.
    
    Args:
        api_url (str): The URL of the API endpoint.
        api_key (str): The API key for authentication.
        query (dict): A dictionary containing the query parameters.
        
    Returns:
        dict: The JSON response from the API if successful.
        None: If the request fails.
    """
    try:
        # Set the headers for the request
        headers = {
            "Authorization": f"Bearer {api_key}",
            "Content-Type": "application/json"
        }
        
        # Make the GET request with the query parameters
        response = requests.post(api_url, headers=headers, params=query)
        
        # Check if the request was successful
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Error: {response.status_code} - {response.text}")
            return None
    except Exception as e:
        print(f"An exception occurred: {e}")
        return None

# Example usage
api_url = "https://comm.chatglm.cn/finglm2/api/query"
api_key = "8ae2ae8f4aa44d55a0f86c5809f23cb5"
query = {
    "sql": "SELECT * FROM constantdb.secumain LIMIT 10",
    "limit": 10
}

response = execute_https_api_request(api_url, api_key, query)
if response:
    print(response)
else:
    print("Failed to fetch the API response.")
```
